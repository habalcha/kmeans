```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("Mopsi_Location_Data.csv", sep= " ", names = ['longitude', 'longitude5', 'latitude'])

df = df.loc[:, ['longitude', 'latitude']]

df
```

```python
from sklearn.cluster import KMeans

n_clusters = 10
n_init = 10
max_iter = 100
tol = 0.0001
random_state = 663
kmeans = KMeans(n_clusters=n_clusters, n_init=n_init, max_iter=max_iter, tol=tol, random_state=random_state)
mean_distance = kmeans.fit_transform(df)
plt.figure()
plt.plot(x = mean_distance[0,:], y = mean_distance[:,1])
```

```python
from sklearn.cluster import DBSCAN

max_dist = 7000
hoodsize = 50
cluster = DBSCAN(eps = max_dist, min_samples = hoodsize)
cluster.fit(df)

plt.figure()
plt.scatter(x = 'latitude', y = 'longitude', c = cluster.labels_, data = df)
plt.show()
```
